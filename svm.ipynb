```python
import pandas as pd
```

```python
pd.read_csv("Salary.csv")
```

```python
data =  pd.read_csv("Salary.csv")
```

```python
data
```

```python
independent = data[["YearsExperience"]] 
```

```python
independent
```

```python
dependent = data[["Salary"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.40,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.svm import SVR
```

```python
Regressor = SVR(kernel = 'poly', gamma = 'scale' , C = 100 )
```

```python
Regressor.fit(x_train,y_train)
              
```

```python
Regressor.n_support_
```

```python
Regressor.intercept_
```

```python
y_predict = Regressor.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test,y_predict)
```

```python
r_score
```

```python
import pickle
```

```python
filename = "Support_Vector_Machine.sav"
```

```python
pickle.dump(Regressor,open(filename,'wb'))
```
